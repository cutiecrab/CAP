```python
import pandas as pan

GJnum = 240001
JNnum = 360001
GJcor = []
JNcor = []

for i in range(127) :
    GJcor.append(GJnum)
    GJnum += 1

GJ = pan.DataFrame(data = GJcor, columns = ['Cor_code'])
GJ['Region'] = 'Gwangju'
GJ.set_index('Cor_code', drop = True, append = False, inplace = True)
GJ.to_csv("./기업/광주/GJ_cor.csv")

        
for i in range(70) :    
    JNcor.append(JNnum)
    JNnum += 1


JN = pan.DataFrame(data = JNcor, columns = ['Cor_code'])
JN['Region'] = 'Jeonnam'
JN.set_index('Cor_code', drop = True, append = False, inplace = True)
JN.to_csv("./기업/전남/JN_cor.csv")

JN
```

```python
Total = pan.concat([GJ,JN])
Total.to_csv("./기업/CorTotal.csv")

Total
```
